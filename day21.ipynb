```python
# Advent of Code 2023
# Day 21 Problem 1
import re
from collections import Counter
import functools

with open("aoc_21_input.txt") as f:
    A = f.read().strip().split("\n")

# c+p test case here
TEST = """...........
.....###.#.
.###.##..#.
..#.#...#..
....#.#....
.##..S####.
.##..#...#.
.......##..
.##.#.####.
.##..##.##.
...........""".strip().split("\n")

for i, line in enumerate(TEST):
    print(f"{i}:\t{line}")
```

```python
def parse(inp):
    sx, sy = -1, -1
    for i,row in enumerate(inp):
        print(row)
        if "S" in row:
            sx = i
            sy = row.index("S")
            inp[i] = row.replace("S", ".")
    return inp, (sx,sy)

def printGarden(G, points):
    for i, row in enumerate(G):
        curr = [ch if (i,j) not in points else "O" for j,ch in enumerate(row)]
        print("".join(curr))
    print()

def p1(inp, steps):
    garden, start = parse(inp)
    H, W = len(garden), len(garden[0])
    currPos = set()
    currPos.add(start)

    for _ in range(steps):
        nextPos = set()
        for (x,y) in currPos:
            for dx,dy in [(-1,0), (1,0), (0,-1), (0,1)]:
                xx = x + dx
                yy = y + dy
                if 0 <= xx <= H -1 and 0 <= yy <= W -1 and garden[xx][yy] != "#":
                    nextPos.add((xx, yy))
        currPos = nextPos
        #print(currPos)
        #printGarden(garden, currPos)
    return len(currPos)
```

```python
p1(TEST, 6)
```
